```python
import sounddevice as sd
from scipy.io.wavfile import write
import matplotlib.pyplot as plt
import numpy as np
```

```python
fs = 8000
duration = 20

myrecording = sd.rec(int(duration*fs), samplerate=fs, channels = 2)
print('Start speaking')
sd.wait()
print('Stop speaking')
```

```python
write('sonja.wav', fs, myrecording)
```

```python
plt.plot(np.linspace(0,20,fs*duration), myrecording[:,0])
plt.show()
```
